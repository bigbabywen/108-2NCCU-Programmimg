### *從PTT爬出貼文之標題、作者、日期等資訊

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.ptt.cc/bbs/DC_SALE/index.html" #此處以ptt的DC_sale版為例
response = requests.get(url) #仿造瀏覽器發出 HTTP GET 方法來瀏覽網頁，取得網址所在頁面的內容
html_txt = response.text #取得網頁原始碼
soup = BeautifulSoup(html_txt,'html.parser')
articles = soup.find_all('div','r-ent')
for article in articles:
    post_title = article.find('div','title').find('a').string #在div title裡包含了<a></a>
    post_authors = article.find('div','author').string
    post_date = article.find('div','date').string
    post_pushs= article.find('div','nrec').string
    print("標題:%s\n作者:%s\n貼文日期:%s\n推文數:%s" % (post_title, post_authors, post_date, post_pushs))
    print("-----------------------------------")
```
